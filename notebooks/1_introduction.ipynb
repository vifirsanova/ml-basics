**Машинное обучение** -- разновидность программирования, цель которой научить вычислительное устройство выявлять скрытые признаки классов, объектов, данных и т.д.

Чтобы научить машину выявлять скрытые признаки, нам нужна **репрезентативная выборка**, т.е. датасет (набор данных)

**Генерализация** -- основная задача машинного обучения

**Bias** -- предвзятость искусственного интеллекта, которая вызвана недостатком данных в выборке либо переизбытком стереотипных суждений в данных

```python
import torchaudio
import matplotlib.pyplot as plt

def load_sample():
    librispeech_data = torchaudio.datasets.LIBRISPEECH(root='./', url='test-clean', download=True)
    waveform, sample_rate, utterance, _, speaker_id, chapter_id = librispeech_data[0]

    print(f"Speaker ID: {speaker_id}, Chapter ID: {chapter_id}")
    print(f"Transcript: {utterance}")
    print(f"Sample Rate: {sample_rate}")
    print(f"Waveform Shape: {waveform.shape}")

    return waveform, sample_rate

def plot_waveform(waveform, sample_rate):
    plt.figure(figsize=(10, 4))
    plt.plot(waveform.t().numpy())
    plt.title(f"Waveform at {sample_rate} Hz")
    plt.show()

waveform, sample_rate = load_sample()
plot_waveform(waveform, sample_rate)
```

```python
from IPython.display import Audio

Audio(waveform.numpy(), rate=sample_rate)
```

Обучение с учителем:
1. есть образцы решения задач
2. эти образцы решения мы называем **метками** (labels)

На интуитивном уровне:
1. **обучение**: мы *скармливаем* машине тексты и метки
2. **обучение (подбор параметров)**: машина ищет скрытые признаки, чтобы научиться связывать тексты с метками
3. **валидация, тестирование, использование**: мы *скармливаем* машине только тексты, чтобы она предсказала только метки

**Параметры** -- это скрытые признаки, которые ищет модель, чтобы сопоставить данные меткам

**Проблема черного ящика** -- это проблема того, что мы до конца не понимаем, как работают модели машинного обучения, потому что параметры закодированы математически, с помощью теории вероятностей

Параметры внутри модели -- это просто числа, которые ничего не значат без специальной математической интерпретации

**Обучение с подкреплением** -- распространено в робототехнике, играх, индустриальные задачи (роботы на заводах)

Например, чем ближе змейка к яблочку, тем больше мы ее награждаем; чем дальше от яблочка, тем сильнее наказываем
